# 1. Introduction

LangChain đơn giản hóa mọi giai đoạn của vòng đời ứng dụng LLM:
- Development: Xây dựng ứng dụng của bạn bằng các thành phần nguồn mở của LangChain và tích hợp của bên thứ ba. Sử dụng LangGraph để xây dựng các tác nhân có trạng thái với hỗ trợ phát trực tuyến và vòng lặp con người hạng nhất.
- Productionzation: Sử dụng LangSmith để kiểm tra, giám sát và đánh giá các ứng dụng của bạn, để bạn có thể liên tục tối ưu hóa và triển khai một cách tự tin.
- Deployment: Biến các ứng dụng LangGraph của bạn thành API và Trợ lý sẵn sàng đưa vào sản xuất với Nền tảng LangGraph.



LangChain triển khai giao diện chuẩn cho các mô hình ngôn ngữ lớn và các công nghệ liên quan, chẳng hạn như mô hình nhúng và kho lưu trữ vector, và tích hợp với hàng trăm nhà cung cấp.

```python
import os
import getpass
from langchain.chat_models import init_chat_model

# GROQ_API_KEY = 'xxx'
if not os.environ.get('GROQ_API_KEY'):
    os.environ['GROQ_API_KEY'] = getpass.getpass('Enter the Groq API KEY: ')

model = init_chat_model('llama3-8b-8192', model_provider='groq')
```

```python
model.invoke('Hello world!')
```

Kiến trúc Khung LangChain bao gồm nhiều thư viện nguồn mở. Đọc thêm trong trang Kiến trúc.
- langchain-core: Các trừu tượng cơ sở cho các mô hình trò chuyện và các thành phần khác.
- Các gói tích hợp (ví dụ: langchain-openai, langchain-anthropic, v.v.): Các tích hợp quan trọng đã được chia thành các gói nhẹ được nhóm LangChain và các nhà phát triển tích hợp đồng bảo trì.
- langchain: Các chuỗi, tác nhân và chiến lược truy xuất tạo nên kiến ​​trúc nhận thức của ứng dụng.
- langchain-community: Các tích hợp của bên thứ ba được cộng đồng bảo trì.
- langgraph: Khung điều phối để kết hợp các thành phần LangChain thành các ứng dụng sẵn sàng sản xuất với tính bền bỉ, phát trực tuyến và các tính năng chính khác. Xem tài liệu LangGraph.
